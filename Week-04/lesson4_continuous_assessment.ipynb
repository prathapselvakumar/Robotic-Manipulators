## Task A

Consider the following values obtained from your row in the spreadsheet.
- $q_{A0}$ is given in degrees by `q_A0`. **Convert them to radians before using them in your answer**.
- $q_{A1}$ is given in metres by `q_A1`. Do not convert this value.

Suppose that we have a `RP` robot defined by the following transformations in SE(2)
- a rotation of $q_{A0}$ about the base frame.
- a translation of $q_{A1}$ about the $x\text{-axis}$ of the current frame.

Find the Jacobian $\mymatrix{J}_A$ that satisfies

$$\left[\begin{array}{c}
         \dot{p}_{x} \\
         \dot{p}_{y} \\
         \dot{\phi}
        \end{array}\right] = \mymatrix{J}_A \left[\begin{array}{c}
         \dot{q}_{A0} \\
         \dot{q}_{A1}
        \end{array}\right]$$

where $p_x$, $p_y$, and $\phi$ are the end-effector $x\text{-axis}$ position, $y\text{-axis}$ position, and rotation angle, respectively.

Compute the value of the Jacobian at $q_{A0}$ and $q_{A1}$. Store the result in `J_A`. The answer must be of type `np.array`.

```python
# Solve the task here.

# Given values
q_A0_deg =-50
q_A1 =0.8

q_A0 = np.radians(q_A0_deg)

# Add your final answer below. Replace the `None`.
J_A = np.array([
    [-q_A1 * np.sin(q_A0), np.cos(q_A0)],
    [ q_A1 * np.cos(q_A0), np.sin(q_A0)],
    [                 1.0,           0.0]
])


print(J_A)



```

## Task B

Consider the following values obtained from your row in the spreadsheet.
- $q_{B0}$ and $q_{B1}$ are given in degrees by `q_B0` and `q_B1` respectively. **Convert them to radians before using them in your answer**.

Given the Jacobian

$$\mymatrix{J}_B = \left[\begin{array}{ccc}
         -\sin{q_{B0}} - \sin{(q_{B0} + q_{B1})} & -\sin{(q_{B0} + q_{B1})} \\
         \cos{q_{B0}} + \cos{(q_{B0} + q_{B1})} & \cos{(q_{B0} + q_{B1})}  \\
         1                 & 1
        \end{array}\right].$$

When the manipulator in the configuration given by $q_{B0}$ and $q_{B1}$, compute the task-space velocity vector, $\dot{\myvec{x}}_B$, using $\mymatrix{J}_B$ and

$$\dot{\myvec{q}} = \left[\begin{array}{c}
         \dot{q}_{B0} \\
         \dot{q}_{B1}
        \end{array}\right] = \left[\begin{array}{c}
         0.1 \\
         0.3
        \end{array}\right].$$

Store the result in `x_dot_B`. The answer must be of type `np.array`.



```python
# Do not modify this(these) value(s). The original value(s) will be used for marking.
q_dot_B0 = 0.1 # Do not convert this value
q_dot_B1 = 0.3 # Do not convert this value
# Solve the task here.

# Given values
q_B0_deg = 80
q_B1_deg = -80

q_B0 = np.radians(q_B0_deg)
q_B1 = np.radians(q_B1_deg)

J_B = np.array([
    [-np.sin(q_B0) - np.sin(q_B0 + q_B1), -np.sin(q_B0 + q_B1)],
    [ np.cos(q_B0) + np.cos(q_B0 + q_B1),  np.cos(q_B0 + q_B1)],
    [                                1.0,                  1.0]
])


q_dot_B = np.array([q_dot_B0, q_dot_B1])

# Add your final answer below. Replace the `None`.
x_dot_B = J_B @ q_dot_B


print("J_B =\n", J_B)
print("x_dot_B =", x_dot_B)


```
