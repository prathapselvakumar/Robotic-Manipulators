## Task A

Use the $\theta_{A1}$ and $\theta_{A2}$ from your row in the spreadsheet.

**In the spreadsheet, those parameters are written in degrees. Convert them to radians before using trigonometric functions.**

Compute the result of the following **_sequential_** rotations in SO(3) starting at the World frame.
- A rotation of $\theta_{A1}$ about the $x-$axis of the _World frame_.
- A rotation of $\theta_{A2}$ about the _current_ $z-$axis.

Store the result in `R_A`, as a `numpy` matrix, representing the SO(3) result.


```python
# Solve the task here.

# Replace these values with the ones from your spreadsheet row!
theta_A1_deg = -50 # Theta A1 in degrees
theta_A2_deg = 30 # Theta A2 in degrees

# Convert to radians
theta_A1 = theta_A1_deg * pi / 180.0
theta_A2 = theta_A2_deg * pi / 180.0

# Define rotation matrices
# Rotation about World x-axis by theta_A1
R_x = np.array([
    [1, 0, 0],
    [0, cos(theta_A1), -sin(theta_A1)],
    [0, sin(theta_A1), cos(theta_A1)]
])

# Rotation about Current z-axis by theta_A2
R_z = np.array([
    [cos(theta_A2), -sin(theta_A2), 0],
    [sin(theta_A2), cos(theta_A2), 0],
    [0, 0, 1]
])


R_A = R_x @ R_z

print(R_A)
```

## Task B

Use $t$ as-is in the translation and rotation below. **Do not** convert $t$.

Consider the translation

$$\myvec{p}_B(t) = \left[\begin{array}{ccc}
         t \\
         t + 4 \\
         t - 2
        \end{array}\right],$$

and the rotation

$$\mymatrix{R}_B(t) = \left[\begin{array}{ccc}
             \cos{t} & -\sin{t} & 0\\
             \sin{t} &  \cos{t} & 0\\
             0       &  0       & 1
        \end{array}\right].$$

Compute the homogenous transformation in 3D, i.e. the SE(3) element, representing the rotation $\mymatrix{R}_B(t)$ followed by the translation $\myvec{p}_B(t)$ when $t=0.5$.

Store the result in `H_B`, as a `numpy` matrix, representing the SE(3) result.

```python
# Do not modify this(these) value(s). The original value(s) will be used for marking.
t = 0.5
# Solve the task here.

# Rotation matrix R_B(t)
R_B = np.array([
    [cos(t), -sin(t), 0],
    [sin(t), cos(t), 0],
    [0, 0, 1]
])

# Translation vector p_B(t)
p_B = np.array([t, t + 4, t - 2])

# Construct Homogeneous Transformation Matrix H_B (4x4)
# H_B = [[R, p], [0, 1]]
H_B = np.eye(4)
H_B[0:3, 0:3] = R_B
H_B[0:3, 3] = p_B

print(H_B)
```
