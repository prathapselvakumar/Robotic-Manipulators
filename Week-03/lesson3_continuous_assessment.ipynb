## Task A

| Joint | $\theta$                | $d$      | $a$      | $\alpha$    |
|-------|-------------------------|----------|----------|-------------|
| 1     | $q_{A0}(t) + \phi_{A0}$ | $l_{A0}$ | 0        | 0           |
| 2     | $q_{A1}(t)$             | 0        | $l_{A1}$ | $\phi_{A1}$ |

For the DH-parameter table above, consider the following values obtained from your row in the spreadsheet.
- $\phi_{A0}$ and $\phi_{A1}$ are given in degrees by `Ď•_A0` and `Ď•_A1` respectively. **Convert them to radians before using them in your answer**.
- $l_{A0}$ and $l_{A1}$ are given in metres by `l_A0` and `l_A1` respectively. Use them as-is.

Calculate the forward kinematics of this manipulator, given by
$$\mymatrix{H}^{A0}_{A2} \in SE(3),$$
using the DH parameter table above.

Compute the FKM at $q_{A0} = \frac{\pi}{4}$ and $q_{A1} = -\frac{\pi}{2}$ and store it in `H_A0_A2` of the cell below. The answer must be of type `np.array`.


```python
# Do not modify this(these) value(s). The original value(s) will be used for marking.
q_A0 = pi/4 # Do not convert this value
q_A1 = -pi/2 # Do not convert this value
# Solve the task here.



phi_A0_deg = -50
phi_A1_deg = 30
l_A0 = 0.8
l_A1 = 0.3

# Given joint values
q_A0 = pi/4
q_A1 = -pi/2

# Convert degrees to radians
phi_A0 = phi_A0_deg * pi / 180
phi_A1 = phi_A1_deg * pi / 180

# DH parameters
theta1 = q_A0 + phi_A0
d1 = l_A0
a1 = 0
alpha1 = 0

theta2 = q_A1
d2 = 0
a2 = l_A1
alpha2 = phi_A1

def DH(theta, d, a, alpha):
    return np.array([
        [cos(theta), -sin(theta)*cos(alpha),  sin(theta)*sin(alpha), a*cos(theta)],
        [sin(theta),  cos(theta)*cos(alpha), -cos(theta)*sin(alpha), a*sin(theta)],
        [0,           sin(alpha),             cos(alpha),            d],
        [0,           0,                      0,                     1]
    ])


# Add your final answer below. Replace the `None`.
H_A0_A2 = DH(theta1,d1,a1, alpha1) @ DH(theta2, d2, a2, alpha2)

print(H_A0_A2)
```
